### Dataset cleaning process

Some transformations and cleaning techniques were applied to the dataset so as to make it functional to work with. Firstly, useless columns were deleted using `.drop()`:

    columns_to_delete = ["Case Number", "pdf", "href formula", "href", "Case Number.1", "Case Number.2", "original order", "Unnamed: 22", "Unnamed: 23"]
    df = df.drop(columns_to_delete, axis=1)
    
Then I checked for rows filled with NaN's and deleted them using the reverse `~` to both `isna()` and `.all()`:

    df = df[~df.isna().all(1)]
    
Then, all elements in 'Country' were transfored into uppercase so they all had the same format:
    
    `df['Country'] = df['Country'].str.upper()`
    
Using RegEx, a pattern was searched in the date column to create a new 'month' column:
    
    df["Month"] = df.Date.apply(lambda x: "".join(re.findall(r"[A-Za-z]{3}-",x)).replace("-",""))
    
The right space from the name of the column 'Sex' was stripped as well as from its values. Then the column "Sex " with a space was dropped:

    df['Sex'] = df['Sex '].str.rstrip()
    df = df.drop("Sex ", axis=1)
    
Then I looked for elements with two digits in 'Age' to clean the column:
 
    df['Age'] = df['Age'].dropna().apply(lambda x: re.findall(r"\d{2}",x))  
    df['Age'] = df['Age'].str[0] #Â Dropping extra values in Age (sometimes more than one in one row
    df['Age'] = df['Age'].astype(float) #Convert all values into floats so we can later operate with them
    
Finally, the spaces left and right from the 'Fatal (Y/N)' column were stripped, converted all the elements into uppercase, replaced invalid values by 'UNKNOWN' and filled NaNs with 'UNKNOWN'as well . Then, I creted a new column 'Fatal' and dropped the'Fatal (Y/N)' column:

    df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.rstrip().str.lstrip().str.upper()
    df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace({'M': 'UNKNOWN','2017':'UNKNOWN'}).fillna("UNKNOWN") #dealing with NaNs    
    df['Fatal'] = df['Fatal (Y/N)']
    df = df.drop("Fatal (Y/N)", axis=1)

## Conclusions

In absolute terms, shark attacks are a rare event, with a rate of 47 attacks per year from 1900 to 2018 (see operation below). Yet, it is a risk that must be taken seriously when organizing water activities such as scuba diving immersions. 

Therefore, after conducting an analysis of the Shark Attacks database, these are some recomendations to an international water sports and expereriences companyfor increasing the safety of their scuba diving activities:
1. In terms of location, try to avoid the **USA, Australia and South Africa**.
    - They combine for a total of 4146 attacks, that is, almost 66% of the attacks happened in these three countries.
    - Philipines, Egypt or Croatia could be alternative safer options, but still highly popular touristically speaking
2. Shark attacks have been increasing steadily and 2015 was the year in which most attacks were registered with 143. Nevertheless, since that year there was a sharp drop, with 'just' 53 attacks recorded in 2018. Hence, statistically there should not be worry of an increasing in attacks per year.
3. Swimming, surfing and fishing are the activities during which more attacks happened. Only 76 attacks ocurred while the victim was scuba diving, which accounts for little over 1% of the attacks. Consequently, it can be stated that it will be safe for the company to implement a scuba diving experience.
4. Regarding the months in which attacks are less frequent, if we exclude the countries mentioned in the first bullet point, january to may is the safest period of the year according to the data.
